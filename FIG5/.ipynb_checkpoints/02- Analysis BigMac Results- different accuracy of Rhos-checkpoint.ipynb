```python
# estimated F1
for c in errors_in_roh:
    with open(address_main+ 'Iric, c=' + ' ' + str(c)+'.pickle' , 'rb') as handle:
        [dic_inversions_F1_,dic_inversions_LL] = pickle.load(handle)
        
    # make Averages 
    inversions_AVG = []
    for rho in dic_inversions_F1_.keys():
        inversions_AVG.append(np.mean(dic_inversions_F1_[rho]))
    
    
    
    if c == 1:
        linestyle = '-'
        print(linestyle)
    else:
        linestyle = ':'
    
    
    l = [x * 100 for x in inversions_AVG]
    plt.plot(np.arange(0.02,1.02,0.02), l , c='b', linestyle = linestyle)
    
    # Text (c)
    if c > 1:
        temp = 0.1
    elif c < 1:
        temp = 0.00
    else:
        temp = 0.03
    
    plt.text(x= 1 + temp  
             ,y=l[-1]
             ,s=str(c),color='b',fontsize=14)
    
plt.grid(axis='both')
plt.xlim([0,1.2])
plt.ylim([-0.1,8])
plt.text(x = 0.6,y= 6.7, s= "estimated F1", color = 'b', fontsize=14)
plt.text(x = 0.6,y= 6.0, s= "LL-score", color = 'r', fontsize=14)
plt.text(x = 1.07,y= 4.3, s= "c", color = 'b', fontsize=18)
plt.savefig(address_main+'inversion_VS_estimated_rho(Iris).png')
plt.show()
```
